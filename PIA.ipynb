```python
class Asistente:
    def __init__(self, matricula, nombre, apellido1, apellido2, fecha_nacimiento, carrera):
        self.matricula = matricula
        self.nombre = nombre
        self.apellido1 = apellido1
        self.apellido2 = apellido2
        self.fecha_nacimiento = fecha_nacimiento
        self.carrera = carrera

class RegistroEvento:
    def __init__(self, asistente, evento, auditorio, fecha, hora):
        self.asistente = asistente
        self.evento = evento
        self.auditorio = auditorio
        self.fecha = fecha
        self.hora = hora

class Acciones:
    asistentes = []
    registros_evento = []

    @classmethod
    def agregar_asistente(cls, asistente):
        cls.asistentes.append(asistente)

    @classmethod
    def eliminar_asistente(cls, matricula):
        cls.asistentes = [a for a in cls.asistentes if a.matricula != matricula]

    @classmethod
    def actualizar_asistente(cls, matricula, nuevo_asistente):
        for i, a in enumerate(cls.asistentes):
            if a.matricula == matricula:
                cls.asistentes[i] = nuevo_asistente
                break

    @classmethod
    def mostrar_info_asistente(cls, matricula):
        for a in cls.asistentes:
            if a.matricula == matricula:
                return f"Matrícula: {a.matricula}\nNombre: {a.nombre} {a.apellido1} {a.apellido2}\nFecha de Nacimiento: {a.fecha_nacimiento}\nCarrera: {a.carrera}"
        return "Asistente no encontrado."

    @classmethod
    def registrar_asistente_evento(cls, asistente, evento, auditorio, fecha, hora):
        if cls.capacidad_disponible(auditorio) > 0:
            registro = RegistroEvento(asistente, evento, auditorio, fecha, hora)
            cls.registros_evento.append(registro)
            return "Registro exitoso."
        else:
            return "El auditorio está lleno."

    @classmethod
    def eliminar_registro_asistente_evento(cls, matricula, evento):
        cls.registros_evento = [r for r in cls.registros_evento if r.asistente.matricula != matricula or r.evento != evento]

    @classmethod
    def mostrar_info_registro_asistente(cls, matricula):
        info_registros = []
        for r in cls.registros_evento:
            if r.asistente.matricula == matricula:
                info_registros.append(f"Evento: {r.evento}\nAuditorio: {r.auditorio}\nFecha y Hora: {r.fecha} {r.hora}")
        if info_registros:
            return "\n".join(info_registros)
        else:
            return "No hay registros para este asistente."

    @classmethod
    def capacidad_disponible(cls, auditorio):
        capacidad_total = {"A": 1000, "B": 200, "C": 150}
        asistentes_en_auditorio = sum(1 for r in cls.registros_evento if r.auditorio == auditorio)
        return capacidad_total[auditorio] - asistentes_en_auditorio

class AccionesGenerales(Acciones):
    @classmethod
    def guardar_datos(cls):
       
        print("Datos guardados.")

    @classmethod
    def leer_datos(cls):
         
        print("Datos leídos.")

    @classmethod
    def mostrar_espacios_disponibles_por_conferencia(cls):
        for auditorio in ["A", "B", "C"]:
            capacidad_disponible = cls.capacidad_disponible(auditorio)
            print(f"Auditorio {auditorio}: {capacidad_disponible} asientos disponibles.")

    @classmethod
    def mostrar_lista_eventos_registrados_para_asistente(cls, matricula):
        info_registros = cls.mostrar_info_registro_asistente(matricula)
        if info_registros != "No hay registros para este asistente.":
            print(f"Eventos registrados para el asistente {matricula}:\n{info_registros}")
        else:
            print(info_registros)

    @classmethod
    def emitir_constancia_participacion(cls, matricula):
        eventos_asistidos = [r.evento for r in cls.registros_evento if r.asistente.matricula == matricula]
        if len(eventos_asistidos) >= 3:
            print(f"Constancia de participación emitida para el asistente {matricula}.")
        else:
            print(f"El asistente {matricula} no cumple con los requisitos para la constancia de participación.")


asistente1 = Asistente("A12345", "Juan", "Gómez", "Pérez", "2000-01-15", "LTI")
asistente2 = Asistente("A67890", "María", "López", "Martínez", "1999-05-20", "LA")

Acciones.agregar_asistente(asistente1)
Acciones.agregar_asistente(asistente2)

AccionesGenerales.guardar_datos()
AccionesGenerales.leer_datos()

Acciones.registrar_asistente_evento(asistente1, "Inteligencia Artificial en los Negocios", "A", "2023-11-04", "15:00")
Acciones.registrar_asistente_evento(asistente1, "Uso de la nube para gestión de procesos", "B", "2023-11-05", "09:00")

AccionesGenerales.mostrar_espacios_disponibles_por_conferencia()
AccionesGenerales.mostrar_lista_eventos_registrados_para_asistente("A12345")

Acciones.eliminar_asistente("A12345")

AccionesGenerales.mostrar_espacios_disponibles_por_conferencia()
AccionesGenerales.mostrar_lista_eventos_registrados_para_asistente("A12345")

AccionesGenerales.emitir_constancia_participacion("A67890")
AccionesGenerales.emitir_constancia_participacion("A12345")
```
